```python
#calculator
import tkinter
from tkinter import *
obj =Tk()
obj.geometry("500x600")
obj.title("calculator")

label=Label(obj,text="Calculator",font="arial 30",fg="red")
label.pack()
label_2=Label(obj,text="Calculation",font="arial 15",fg="red")
label_2.pack()
def click(event):
    text=event.widget.cget("text")
    print(text)
    if text=="=":
        if value.get().isdigit():
            display=int(value.get())
        else:
            display=eval(screen.get())
        value.set(display)
        screen.update()
    elif text=="c":
        value.set("")
        screen.update()
    else:
        value.set(value.get() + text)
        screen.update()
value=StringVar()
value.set("")
screen=Entry(obj,font=45,textvariable=value)
screen.pack(fill=X,ipadx=8,pady=10,padx=10)

f=Frame(obj,bg="grey")
b=Button(f,text="1",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="2",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="3",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
f.pack()

f=Frame(obj,bg="grey")
b=Button(f,text="4",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="5",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="6",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
f.pack()

f=Frame(obj,bg="grey")
b=Button(f,text="7",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="8",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="9",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
f.pack()

f=Frame(obj,bg="grey")
b=Button(f,text="0",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="=",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="c",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
f.pack()

f=Frame(obj,bg="grey")
b=Button(f,text="*",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="/",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="+",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
f.pack()

f=Frame(obj,bg="grey")
b=Button(f,text="-",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
b=Button(f,text="%",font=60,padx=20,pady=20)
b.pack(side=LEFT)
b.bind("<Button-1>",click)
f.pack()

obj.mainloop()
```
